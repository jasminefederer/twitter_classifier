```python
import tweepy
import json
import requests
from nltk.corpus import stopwords # stopword examples, 'its', 'on', 'the', etc <---- will be helpful later
# most pythonistas will rename pandas as pd, numpy as np, and datetime as dt for short (you don't have to)
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# open file with keys and set the path to your credentials JSON file
# you'll need to replace my file with yours
credentials = "twitter_keys.json"
with open(credentials, "r") as keys:
    api_tokens = json.load(keys)
```

```python
bearer_token = api_tokens["bearer_token"]
```

```python
# we only want status info so we won't need the higher level access

client = tweepy.Client(bearer_token=bearer_token,return_type = tweepy.Response)
```

```python
# first classifier
# replace 'tory' with whatever you would like to search for
# if you want to search for a term with multiple words, use _ in place of space

tweets_tory = client.search_recent_tweets( # instead of searching for most recent tweets you could instead search for a specific date/time frame
# this could be useful for journalistic purposes to see how opinions and data on certain topics/trends has changed over time
# ... or to snapshot tweets from a particular date in history, e.g. Trump inauguration
    query = "tory -is:retweet",  # searches for #tory while ignoring retweets
    max_results = 100
)

tweets_tory.data
```

```python
list_of_tory_tweets = []
for tweet in tweets_tory.data:
    list_of_tory_tweets.append(tweet.text)
    
list_of_tory_tweets
```

```python
import nltk
from nltk.corpus import stopwords
import random

def get_features(text):
    # create array of keywords
    
    # make text lowercase
    text = text.lower()
    
    # split text into an array of words
    terms = text.split()
    
    # strip away punctuation from tweets
    terms = text.strip()

    s_words = stopwords.words('english')
    # this following part isn't necessary but will make the next part shorter
    # could also look into pip install clean-text to eliminate emojis
    
    keywords = [term for term in terms if term not in s_words and '@' not in term] # removing @ will take out usernames
     # could also look into pip install clean-text to eliminate emojis as well as strip function
    random.shuffle(keywords) # see note below on the use of this
    while len(keywords) < 3:
        keywords.append('None')
    
    return {'first_keyword': keywords[0], 'second_keyword': keywords[1], 'third_keyword': keywords[2]}
# feel free to change and experiment with these features
# i.e. collections library counter func might be useful to look into if you want to find the most used words
# if you took out random.shuffle line then it would display the keywords which come first, second, third chronologically
```

```python
# second classifier
# replace 'evil' with whatever you would like to search for

tweets_evil = client.search_recent_tweets(
    query = "evil -is:retweet",  # searches for #evil while ignoring retweets
    max_results = 100
)

tweets_evil.data
```

```python
list_of_evil_tweets = []
for tweet in tweets_evil.data:
    list_of_evil_tweets.append(tweet.text)
    
list_of_evil_tweets
```

```python
# create a list of tuples containing the featureset and the label
# features and label list will contain my features and my label

features_and_labels = []

for tweet in list_of_tory_tweets:
    features_and_labels.append((get_features(tweet), 'tory'))
```

```python
for tweet in list_of_evil_tweets:
    features_and_labels.append((get_features(tweet), 'evil'))
```

```python
# mixes all evil and tory tweets together
random.shuffle(features_and_labels)
```

```python
# Bayes Theoreum
# split 80:20 training data and testing data

import math

# get how many elements in my list is 80%
split_num = math.floor(len(features_and_labels) * .8)
split_num
```

```python
# extracting 80% of my data for my training set
training_set = features_and_labels[:split_num]

# remaining 20% for testing
testing_set = features_and_labels[split_num:]
```

```python
import nltk
# create our classifier
# baseline is 50% because I have two labels/outcomes
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
# include a term or full tweet that you would like the algorithm to try and classify

classifier.classify(get_features("soulless"))
```

```python
# test accuracy
nltk.classify.accuracy(classifier, testing_set)
```
